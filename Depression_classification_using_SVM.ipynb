```python
combined_data = pd.concat([new_nondep_data, new_dep_data], ignore_index=True)
print(combined_data.head())
print(combined_data.tail())
```

```python
import re
import string
def clean_text(text):
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\s+', ' ', text.strip())
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    return text
combined_data['cleaned_tweet'] = combined_data['tweet'].apply(clean_text)
```

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python

nltk.download('punkt')        # tokenizer
nltk.download('punkt_tab')    # the one your error mentioned
nltk.download('wordnet')      # lemmatizer
nltk.download('omw-1.4')      # wordnet data
nltk.download('stopwords')    # if you use stopwords
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word not in stopwords.words('english') and word.isalpha()]
    stemmed = [stemmer.stem(word) for word in tokens]
    lemmatized = [lemmatizer.lemmatize(word) for word in stemmed]
    return ' '.join(lemmatized)
combined_data['fin_tweet'] = combined_data['cleaned_tweet'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(combined_data['fin_tweet'])
Y = combined_data['Y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
pca = PCA(n_components=100)  #Can be commented out
X_train_reduced = pca.fit_transform(X_train.toarray())#Can be commented out
X_test_reduced = pca.transform(X_test.toarray())#Can be commented out
```

```python
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_reduced, y_train)#if be commented out .X_train_reduced->X_train
```

```python
from sklearn.metrics import accuracy_score
y_pred = svm_model.predict(X_test_reduced)#if be commented out .X_test_reduced->X_test
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```
